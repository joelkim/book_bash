# 환경

## 프로필 관련 파일

홈 디렉토리에 있는 다음 파일들은 사용자의 셸 환경을 설정하는 데 사용됩니다.

- `.bash_profile`
- `.bashrc`
- `.bash_logout`

## `.bash_profile` 파일

- `.bash_profile` 파일은 배시 셸에 로그인 할 때 자동으로 실행되는 파일입니다.

### 대체 파일

- 배시 셸이 아닌 경우 `.bash_profile` 대신 다음과 같은 파일명을 사용합니다.
  - `.bash_login`: C 셸에서 사용하는 파일명
  - `.profile`: Bourne 셸, Korn 셸에서 사용하는 파일명
- 배시 셸에서 `.bash_profile` 파일이 존재하지 않고 대체 파일이 존재하는 경우에는 해당 대체 파일을 `.bash_profile` 파일로 사용합니다.
- 대체 파일을 찾는 순서는 `.bash_profile` -> `.bash_login` -> `.profile` 순서입니다. 
- 즉, `.bash_profile` 파일이 존재하면 다른 파일들은 무시됩니다.
- 만약 `.bash_login`, `.profile` 두 파일이 존재하면 `.bash_login` 파일이 사용되고 `.profile` 파일은 무시됩니다.

## `.bashrc` 파일

- `.bashrc` 파일은 셸 로그인시가 아니라 로그인 후 bash 명령을 실행하거나 서브셸 실행시, 즉 "비로그인" 환경에서 자동으로 실행되는 파일입니다.
- 일반적으로 `.bash_profile` 파일에서 `.bashrc` 파일을 호출하여 로그인 셸과 비로그인 셸 모두에서 동일한 환경 설정을 사용할 수 있도록 합니다.

```bash
if [ -f ~/.bashrc ]; then
    source ~/.bashrc
fi
```

## `.bash_logout` 파일

- `.bash_logout` 파일은 셀 로그아웃시에 자동으로 실행되는 파일입니다.
